## OECD

```python
# Use modified reader function that allows to download only a specific series, whole datasets may be too large.
# To download whole dataset, use, e.g., NAAG/all
# dataset: National Accounts at a Glance
# series: Net lending/net borrowing, general government, percentage of GDP, all countries
datareader = OECDReaderSeries('NAAG/.B9S13S',start='JAN-01-2006')
data_def = datareader.read()
#datareader = OECDReaderSeries('SNA_TABLE1/.B1_GA.G',start='JAN-01-2006')
datareader = OECDReaderSeries('NAAG/.GDPG',start='JAN-01-2006')
data_growth = datareader.read()
```

```python
def_year2009ind = np.where(data_def.index.year == 2009)[0][0]
def_country_names = data_def.columns.droplevel(1)
def_year2009 = data_def.to_numpy()[year2009ind,:]

growth_year2009ind = np.where(data_growth.index.year == 2009)[0][0]
growth_country_names = data_growth.columns.droplevel(1)
growth_year2009 = data_growth.to_numpy()[year2009ind,:]

# merge data for growth and deficits
country_names_set = set(def_country_names).intersection(set(growth_country_names))
data = {'country_names':[], 'deficit':[], 'growth':[]}

for ind,cname in enumerate(country_names_set):
    def_ind = np.where(def_country_names == cname)
    growth_ind = np.where(growth_country_names == cname)
    if not(np.isnan(def_year2009[def_ind])) and not(np.isnan(growth_year2009[growth_ind])):
        data['country_names'].append(cname)
        data['deficit'].append(float(def_year2009[def_ind]))
        data['growth'].append(float(growth_year2009[growth_ind]))

# rename some data
name_switch = [["China (People's Republic of)",'China'],
               ['Slovak Republic','Slovakia'],
               ['Czech Republic','Czechia']]
for n in name_switch:
    for ind,old_name in enumerate(data['country_names']):
        if n[0]==old_name:
            data['country_names'][ind] = n[1]
            
# sort data
args = np.argsort(data['growth'])
data['deficit'] = [data['deficit'][i] for i in args]
data['growth'] = [data['growth'][i] for i in args]
data['country_names'] = [data['country_names'][i] for i in args]

N_data = len(data['country_names'])
loc_data = np.arange(N_data)
   
param = {'title': '','figsize' : [18,9],
         'xlim': [-0.5,N_data-0.5],'ylim': [-16,12],
         'xlabel': '', 'ylabel': 'Deficit in Percent of GDP / Growth in Percent',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': True,
         'showNBERrecessions': False, 'showNBERrecessions_y': [-30,5]}

fig,ax = myGenerateBarPlot(param)
bar_width = 0.4

ax.bar(loc_data-bar_width/2,data['deficit'],bar_width,color=myColor['tolVibrantBlue'],label='2009 General Government Deficit')
ax.bar(loc_data+bar_width/2,data['growth'],bar_width,color=myColor['tolVibrantOrange'],label='2009 GDP Growth')
ax.set_xticks(ticks=np.linspace(0,N_data-1,len(data['country_names'])))
ax.set_xticklabels(data['country_names'],rotation='vertical');
ax.legend(loc='upper left');
```
