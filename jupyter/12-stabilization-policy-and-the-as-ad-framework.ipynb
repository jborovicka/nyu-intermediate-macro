# Taylor rule

```python
# interest rates data
data_i = myLoadDataFRED(series=['FEDFUNDS','GS1','GS10','AAA','BAA'],transform='none')
data_pi = myLoadDataFRED(series=['CPILFESL'],transform='pct_change_year_ago')

param = {'title': '',
         'xlim': [1954,2022], 'ylim': [0,20],
         'xlabel': 'year', 'ylabel': 'percent',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': False,
         'showNBERrecessions': True, 'showNBERrecessions_y': [0,25]}

Rbar = 2
m = 0.5
pibar = 2

# federal funds rate
fig,ax = myGenerateTSPlot(param)

ax.plot(data_i['year'],data_i['FEDFUNDS'],
                    '-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='Effective Federal Funds Rate')
ax.plot(data_pi['year'],data_pi['CPILFESL']+Rbar+m*(data_pi['CPILFESL']-pibar),
                    '-',linewidth=3,marker='',color=myColor['tolVibrantOrange'],label='Taylor rule policy rate')
ax.legend(loc='upper right')
```
