```python
# load and process the spreadsheet
data_Clev_orig = pd.read_excel('https://www.clevelandfed.org/~/media/files/webcharts/crediteasing/crediteasingbalancesheet.xls',sheet_name="Chart Data")
data_Clev_orig.columns = data_Clev_orig.iloc[0,:]
data_Clev_orig = data_Clev_orig.iloc[1:,:]
data_Clev_orig = data_Clev_orig.set_index('Date')
dates = year_frac(data_Clev_orig.index)
T = len(dates)

# keep only relevant columns, and convert to float
data_columns = ['Traditional Security Holdings','Lending to Financial Institutions',
                'Liquidity to Key Credit Markets','Federal Agency Debt and Mortgage-Backed Securities Purchases',
                'Long-Term Treasury Purchases']
data_Clev_orig = data_Clev_orig[data_columns]
data_Clev_orig = data_Clev_orig.astype('float64')

# change index to numerical years if needed
# data_Clev_orig.reset_index(drop=True,inplace=True)
# data_Clev_orig['date_num'] = dates
# data_Clev_orig = data_Clev_orig.set_index('date_num')

param = {'title': '',
         'xlim': [2007,2022.9], 'ylim': [0,9000],
         'xlabel': '', 'ylabel': 'Billions of USD',
         'ylogscale': False,
         'showgrid': True,
         'showNBERrecessions': True, 'showNBERrecessions_y': [0,9000]}

fig,ax = myGenerateTSPlot(param)

import matplotlib.patches as mpatches
cumval = np.zeros(T) + data_Clev_orig[data_columns[0]]*0.0

legend_patches = []
for ind,lab in enumerate(data_columns):
    cumvalinc = cumval + data_Clev_orig[lab].to_numpy()/1000
    ax.fill_between(dates,cumval,cumvalinc,color=clist_1[ind])
    cumval = cumvalinc
    legend_patches.append(mpatches.Patch(color=clist_1[ind],label=data_Clev_orig[lab].name))

ax.legend(handles=legend_patches,loc='upper left');

ax.annotate(text='QE1',xy=(2009,2500),xycoords='data', # arrow tip, alternative xycoords is 'axes fraction'
            xytext=(2009,3500), textcoords='data',# box position, alternative textcoords is 'axes fraction'
            horizontalalignment='center', verticalalignment='bottom',**style)
ax.annotate(text='QE2',xy=(2011.25,3000),xycoords='data', # arrow tip, alternative xycoords is 'axes fraction'
            xytext=(2011.25,4000), textcoords='data',# box position, alternative textcoords is 'axes fraction'
            horizontalalignment='center', verticalalignment='bottom',**style)
ax.annotate(text='QE3',xy=(2013.5,4000),xycoords='data', # arrow tip, alternative xycoords is 'axes fraction'
            xytext=(2013.5,5000), textcoords='data',# box position, alternative textcoords is 'axes fraction'
            horizontalalignment='center', verticalalignment='bottom',**style)
ax.annotate(text='tapering',xy=(2019,4250),xycoords='data', # arrow tip, alternative xycoords is 'axes fraction'
            xytext=(2019,5250), textcoords='data',# box position, alternative textcoords is 'axes fraction'
            horizontalalignment='center', verticalalignment='bottom',**style)
ax.annotate(text='pandemic support',xy=(2020.4,7250),xycoords='data', # arrow tip, alternative xycoords is 'axes fraction'
            xytext=(2020.4,8250), textcoords='data',# box position, alternative textcoords is 'axes fraction'
            horizontalalignment='right', verticalalignment='bottom',**style)


# alternative way of plotting the same data using the data.plot.area function
#ax = data_Clev_orig.plot.area(color=[myColor['tolVibrantBlue'],
#                           myColor['tolVibrantOrange'],
#                           myColor['tolVibrantCyan'],
#                           myColor['tolVibrantMagenta'],
#                           myColor['tolVibrantTeal']],
#                    linewidth=0)
#ax.grid(color=myColor['tolPaleGrey'],linestyle=':')
```
