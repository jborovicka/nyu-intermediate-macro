# U.S. inflation

U.S. inflation data from the <a href="https://fred.stlouisfed.org/">FRED database</a> at the Federal Reserve Bank of St. Louis.

```python
# inflation data over a longer sample
data = myLoadDataFRED(series=['CPIAUCNS','CPILFESL'],transform='pct_change_year_ago')

param = {'title': '',
         'xlim': [1913,2022], 'ylim': [-20,25],
         'xlabel': 'year', 'ylabel': 'Percent Change from Year Ago',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': True,
         'showNBERrecessions': True, 'showNBERrecessions_y': [-20,25]}

fig,ax = myGenerateTSPlot(param)

ax.plot(data['year'],data['CPIAUCNS'],
                    '-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='Consumer Price Index: All Items')
#ax.plot(data['year'],data['CPILFESL'],
#                    linewidth=3,marker='',color=myColor['tolVibrantOrange'],label='Consumer Price Index: All Items Less Food and Energy')
x = ax.legend(loc='upper right')
```

## Longer inflation time series, from MeasuringWorth.com
http://www.measuringworth.com/datasets/uscpi/result.php

```python
# load locally stored data
data_MW = pd.read_csv('data/USCPI_1774-2022.csv',skiprows=3)

param = {'figsize' : [15,9], 'fontsize': 16,
         'title': '',
         'xlim': [1774,2022], 'ylim': [-25,25],
         'xlabel': 'year', 'ylabel': 'inflation rate (%)',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': False,
         'showNBERrecessions': True, 'showNBERrecessions_y': [-25,25]}

fig,ax = myGenerateTSPlot(param)

year = data_MW['Year'].to_numpy()
CPI = data_MW['U.S. Consumer Price Index *'].to_numpy()
inflation = (CPI[1:]-CPI[0:-1])/CPI[1:]*100

ax.plot(year[1:],inflation,'-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='Consumer Price Index: All Items')
```

```python
param = {'figsize' : [15,9], 'fontsize': 16,
         'title': '',
         'xlim': [1774,2022], 'ylim': [5,320],
         'xlabel': 'year', 'ylabel': 'consumer price index',
         'ylogscale': True,
         'showgrid': True, 'highlightzero': False,
         'showNBERrecessions': True, 'showNBERrecessions_y': [5,320]}

fig,ax = myGenerateTSPlot(param)

ax.plot(year,CPI,'-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='Consumer Price Index: All Items')
ax.minorticks_off()   
ax.set_yticks([5,10,20,40,80,160,320]);
```

## Real and nominal interest rates

```python
# inflation data over a longer sample
data = myLoadDataFRED(series=['DGS5','DFII5','T5YIE'],transform='none')

param = {'title': '',
         'xlim': [2002,2022], 'ylim': [-2,6],
         'xlabel': 'year', 'ylabel': 'Percent',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': True,
         'showNBERrecessions': True, 'showNBERrecessions_y': [-2,6]}

fig,ax = myGenerateTSPlot(param)

ax.plot(data['year'],data['DGS5'],
                    '-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='5-Year Treasury Constant Maturity Rate')
ax.plot(data['year'],data['DFII5'],
                    '--',linewidth=3,marker='',color=myColor['tolVibrantOrange'],label='5-Year Treasury Inflation-Indexed Security, Constant Maturity')
x = ax.legend(loc='upper right')
ax.set_xticks([2002,2004,2006,2008,2010,2012,2014,2016,2018,2020,2022]);

param = {'title': '',
         'xlim': [2002,2022], 'ylim': [-3,4],
         'xlabel': 'year', 'ylabel': 'Percent',
         'ylogscale': False,
         'showgrid': True, 'highlightzero': True,
         'showNBERrecessions': True, 'showNBERrecessions_y': [-3,4]}

fig,ax = myGenerateTSPlot(param)

ax.plot(data['year'],data['T5YIE'],
                    '-',linewidth=3,marker='',color=myColor['tolVibrantBlue'],label='5-Year Breakeven Inflation Rate')
x = ax.legend(loc='upper right')
ax.set_xticks([2002,2004,2006,2008,2010,2012,2014,2016,2018,2020,2022]);
```
